```python
#.status_code

#1xx: 정보(Informational)
#클라이언트의 요청이 수신되었으며 처리 중임을 나타냅니다.

#100 Continue: 요청 일부를 성공적으로 수신했으며, 계속 진행하라는 의미입니다.
#101 Switching Protocols: 클라이언트의 프로토콜 변경 요청을 수락했음을 나타냅니다.
#2xx: 성공(Success)
#요청이 성공적으로 처리되었음을 나타냅니다.

#200 OK: 요청이 성공적으로 처리되었으며, 응답 본문에 결과를 포함합니다.
#201 Created: 요청이 성공적으로 처리되었고, 새로운 리소스가 생성되었음을 나타냅니다.
#204 No Content: 요청은 성공했으나, 반환할 내용이 없음을 나타냅니다.
#3xx: 리다이렉션(Redirection)#
# 클라이언트가 요청한 리소스가 다른 위치로 이동했음을 나타냅니다.

#301 Moved Permanently: 요청한 리소스가 영구적으로 새로운 URL로 이동했음을 나타냅니다.
#302 Found: 요청한 리소스가 임시로 다른 URL에 있음(일시적 리다이렉션)을 나타냅니다.
#304 Not Modified: 캐시된 버전이 최신임을 나타냅니다.
#4xx: 클라이언트 오류(Client Error)
#클라이언트가 올바르지 않은 요청을 했음을 나타냅니다.

#400 Bad Request: 잘못된 요청 형식입니다.
#401 Unauthorized: 인증이 필요하거나, 인증에 실패했습니다.
#403 Forbidden: 요청이 서버에 의해 거부되었습니다(권한 문제).
#404 Not Found: 요청한 리소스를 찾을 수 없습니다.
#5xx: 서버 오류(Server Error)
#서버가 요청을 처리하지 못했음을 나타냅니다.

#500 Internal Server Error: 서버에서 예기치 않은 오류가 발생했습니다.
#502 Bad Gateway: 게이트웨이 또는 프록시 서버가 잘못된 응답을 받았습니다.
#503 Service Unavailable: 서버가 과부하 상태이거나 유지보수 중입니다.
#504 Gateway Timeout: 게이트웨이 또는 프록시 서버가 응답을 받는 데 시간이 초과되었습니다.
```
