```python
arr1 = [0] * (9)

i = 1
while i <= 9:
    j = 1
    arr2 = [0] * (9)

    while j <= 9:
        #print(f"{i}*{j}={i*j}")
        arr2[j-1]=f"{i}*{j}={i*j}"
        j = j + 1
    arr1[i - 1] = arr2
    i = i + 1

print(arr1)
print("반복문 종료")
```

```python
arr1 = [0] * (9)

i = 1
while i <= 9:
    if i % 2 == 1:
        j = 1
        arr2 = [0] * (9)

        while j <= 9:
            arr2[j - 1] = i * j
            j = j + 1
        arr1[i - 1] = arr2
    i = i + 1
print(arr1)
print("반복문 종료")
```

```python
i = 1
while i <= 5:
    j = 1
    b = ""
    while j <= 5:
        if i == 3:
            b = b + "O"
        else:
            if i == 1 and j == 3 or i == 5 and j == 3:
                b = b + "O"
            else:
                if i == 2 and 1 < j < 5 or i == 4 and 1 < j < 5:
                    b = b + "O"
                else:
                    b = b + "X"
        j = j + 1
    print(b)
    i = i + 1
```
